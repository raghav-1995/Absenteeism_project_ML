## Using Logistic Regression

### import relevant libraries 

```python
import pandas as pd
import numpy as np
```

```python
## importing preprocessed data
data_preprocessed = pd.read_csv("df_preprocessed.csv")
data_preprocessed.head()
```

### Creating Targets for Logistic Regression

```python
data_preprocessed["Absenteeism Time in Hours"].median()
```

```python
targets= np.where(data_preprocessed["Absenteeism Time in Hours"]>
                data_preprocessed["Absenteeism Time in Hours"].median(),1,0)
targets
```

```python
data_preprocessed["Excessive Absenteeism"]=targets
data_preprocessed.head()
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
data_with_targets.head()
```

```python
## selecting inputs
unscaled_inputs = data_with_targets.iloc[:,:-1]
unscaled_inputs
```

### Standardize The Input

```python
from sklearn.preprocessing import StandardScaler
```

```python
absenteeism_scaler = StandardScaler()
```

```python
unscaled_inputs.columns.values
```

```python
absenteeism_scaler.fit(unscaled_inputs[['Month value', 'Day of week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets']])
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs[['Month value', 'Day of week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets']])
scaled_inputs
```

```python
df_normalized = pd.DataFrame(scaled_inputs,columns=['Month value', 'Day of week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets'])
df_normalized = df_normalized.join(unscaled_inputs[unscaled_inputs.columns.drop(['Month value', 'Day of week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets'])])
columns_reorder = ['reason_type_1', 'reason_type_2', 'reason_type_3', 'reason_type_4',
       'Month value', 'Day of week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets']
df_normalized = df_normalized[columns_reorder]
```

### Split into Train and Test Data's

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_normalized,targets,train_size = 0.8, random_state=20)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

### using Sklearn for Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

### Manually calculating accuracy

```python
model_outputs = reg.predict(x_train)
model_outputs
```

```python
y_train
```

```python
model_outputs == y_train
```

```python
#getting correct output for this many inputs
correct_prediction =np.sum((model_outputs==y_train))
```

```python
# total no of inputs
total=model_outputs.shape[0]
```

```python
manual_accuracy = correct_prediction/total
manual_accuracy
```

### Intercepts and Coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_names=unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns = ["features"],data= feature_names)
summary_table["coefficient"]=np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index+1
summary_table.loc[0]=['Intercept',reg.intercept_[0]]
summary_table
```

```python
summary_table = summary_table.sort_index()
summary_table
```

### Interpreting the coefficient

```python
summary_table["odds_ratio"]=np.exp(summary_table.coefficient)
summary_table.sort_values(['odds_ratio'],ascending = False)
```

### Testing the model

```python
reg.score(x_test,y_test)
```

```python
predict_probalilities = reg.predict_proba(x_test)
predict_probalilities
```

```python
# probabilities of getting output as 1 for each test cases.
predict_probalilities[:,1]
```
